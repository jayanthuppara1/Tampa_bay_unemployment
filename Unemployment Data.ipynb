```python
import pandas as pd

def calculate_weighted_unemployment(unemployment_df, population_df):
    # Initialize a list to store the weighted unemployment rates
    weighted_unemployment = []
    
    # Ensure that 'date' column is in datetime format
    unemployment_df['date'] = pd.to_datetime(unemployment_df['date'])
    population_df['date'] = pd.to_datetime(population_df['date'])
    
    # Extract the year from the 'date' column
    unemployment_df['year'] = unemployment_df['date'].dt.year
    population_df['year'] = population_df['date'].dt.year
    
    # Get unique years from the unemployment DataFrame
    years = unemployment_df['year'].unique()
    
    for year in years:
        # Filter data for the current year
        unemployment_year = unemployment_df[unemployment_df['year'] == year]
        population_year = population_df[population_df['year'] == year]
        
        if population_year.empty or unemployment_year.empty:
            # If there is no data for this year, append NaN
            weighted_unemployment.append(float('nan'))
            continue
        
        # Convert relevant columns to numeric, forcing errors to NaN
        # For each column in unemployment and population data
        unemployment_values = unemployment_year.drop(columns=['date', 'year']).apply(pd.to_numeric, errors='coerce')
        population_values = population_year.drop(columns=['date', 'year']).apply(pd.to_numeric, errors='coerce')
        
        # Calculate total population
        total_population = population_values.sum().sum()
        
        # Calculate weighted average
        # Multiply unemployment rates by populations and sum them
        weighted_sum = (unemployment_values.values * population_values.values).sum()
        
        # Calculate the weighted average unemployment rate
        weighted_avg = weighted_sum / total_population
        weighted_unemployment.append(weighted_avg)
    
    # Return the DataFrame with weighted unemployment data
    return pd.DataFrame({'year': years, 'Tampa Bay Region Unemployment Rate': weighted_unemployment})


```

```python
weighted_unemployment_data = calculate_weighted_unemployment(unemployment_yearly_data, population_data)

```

```python
def format_year_as_date(df):
    # Convert 'year' to string and format as '01/01/YYYY'
    df['date'] = pd.to_datetime(df['year'].astype(str) + '-01-01', format='%Y-%m-%d')
    
    # Drop the original 'year' column if no longer needed
    df = df.drop(columns=['year'])
    
    # Reorder columns to place 'date' first
    df = df[['date', 'Tampa Bay Region Unemployment Rate']]
    
    # Format the 'date' column to MM/DD/YYYY format
    df['date'] = df['date'].dt.strftime('%m/%d/%Y')
    
    return df

# Apply the function to format the 'year' column as date
formatted_data = format_year_as_date(weighted_unemployment_data)

# Display the result
print(formatted_data.head())
```
